## Работаем с полученной в ходе парсинга таблицей

```python
df = pd.read_csv('hh_table_merged.csv')
df
```

*1. У нас есть столбец с заработной платой. Где-то указана только нижняя граница, где-то - диапазон, а где-то - только верхняя. Поскольку мы хотим приблизиться к реальности, из диапазонов мы будем брать нижнюю границу з/п, а там, где указано одно значение - только его. Уберём ненужные символы и преобразуем з/п в числовой формат с помощью функции (экспериментальным методом мы вычислили, что пробелы между разрядами заданы как "\u202f", а также присутствуют зпшки в долларах, их мы умножали на 92):*

```python
def correct(value):
    value = str(value).replace(' ₽', '').replace('от', '').replace('до', '')
    value = value.split(' – ')[0].replace('\u202f', '').replace(' ', '')
    
    if '$' in value:
        value = value.replace('$', '')
        value = int(value)
        value *= 92
    value = int(value)
    if value < 1000:
        value *= 1000
    
    return value
```

```python
df['wage'] = df['wage'].map(correct)
df.head()
```

```python
type(df.wage[2])
```

*2. Теперь у нас в столбике с зарплатами лежат численные значения в рублях. Переведём рейтинг компаний в формат флоат:*

```python
df['company_rate'] = df['company_rate'].map(lambda x: float(str(x).replace(',', '.')))
df.head()
```

*4. Теперь из колонки "employment" сделаем два новых столбца, в которых будут отдельно рабочая занятость и, условно, местоположение работника*

```python
df['employment'] = df['employment'].map(lambda x: x.split(', '))
```

```python
df['working_hours'], df['employee_location'] = df['employment'].map(lambda x: x[0]), df['employment'].map(lambda x: x[1])
```

```python
#Ну и удалим ненужную старую колонку "employment"
df = df.drop('employment', axis=1)
```

```python
df.head()
```

*5. Проверим, какие значения вообще есть в последних двух колонках*

```python
df['working_hours'].unique()
```

```python
df['employee_location'].unique()
```

*Можно увидеть, что в колонке с часами работы 4 уникальных значения, в локации - 5. В последней слова начинаются с маленькой буквы, не совсем очевидно, что такое "полный день", поэтому добавим немного эстетики*

```python
df['employee_location'] = df['employee_location'].map(lambda x: 
                                    'В офисе' if x == 'полный день'
                                    else 'Удалённая работа' if x == 'удаленная работа' 
                                    else 'Сменный график' if x == 'сменный график' 
                                    else 'Гибкий график' if x == 'гибкий график' 
                                    else x)
```
